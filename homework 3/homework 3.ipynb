```python
def sum_distance(from_v, to_v): 
    if from_v > to_v:
        from_v, to_v = to_v, from_v

    summ = 0

    for i in range (from_v, to_v+1):
        summ +=i

    return summ

a = int(input()) #1
b = int(input()) #5

print(sum_distance(a,b))
```

```python
def trim_and_repeat(s, offset, repetitions):
    for rep in range (repetitions):
        s = s[offset:len(s)]
    return s

s = "Hello world! This is homework 3!"
print(trim_and_repeat(s, 0, 1)) 
print(trim_and_repeat(s, 2, 4)) 
print(trim_and_repeat(s, 1, 5)) 
```
